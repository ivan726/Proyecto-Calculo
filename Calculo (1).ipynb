## Solución del Problema

<font size="4"> 
En determinados problemas de termodinámica, el trabajo se podría definir como el área de una determinada región R acotada por las respectivas gráficas de los procesos termodinámicos.
Para la resolución del proble es necesario graficar la región $R$ delimitada por las funciones: <br><br>
$\left\{
    \begin{array}{ll}
        xy = a \\
        xy = b \\
        xy^{1.4} = c \\
        xy^{1.4} = d
\end{array}
\right.$

<br><br>En el siguiente gráfico se mostrará la región $R$:

```python
#Se importan las librerías necesarias para realizar las gráficas...
import matplotlib.pyplot as plt
import numpy as np

#Se insertan los valores de a, b, c, d
a = float(input("Digite la constante a: "))
b = float(input("Digite la constante b: "))
c = float(input("Digite la constante c: "))
d = float(input("Digite la constante d: "))


#Se calculan los puntos de corte entre las gráficas
n1 = (a**3.5)/(c**2.5)      #Punto de corte entre la gráfica xy=a y xy^(1.4)=c
n2 = (a**3.5)/(d**2.5)      #Punto de corte entre la gráfica xy=a y xy^(1.4)=d
n3 = (b**3.5)/(c**2.5)      #Punto de corte entre la gráfica xy=b y xy^(1.4)=c
n4 = (b**3.5)/(d**2.5)      #Punto de corte entre la gráfica xy=b y xy^(1.4)=d


#Estableciendo el intervalo de las Funciones:
x1 = np.arange(n2, n1, 0.01)     #Intervalo de la funcion xy=a: {n2<x<n1}
x2 = np.arange(n4, n3, 0.01)     #Intervalo de la funcion xy=b: {n4<x<n3}
x3 = np.arange(n1, n3, 0.01)     #Intervalo de la funcion xy^(1.4)=c: {n1<x<n3}
x4 = np.arange(n2, n4, 0.01)     #Intervalo de la funcion xy^(1.4)=c: {n2<x<n4}


#Se define la función de la forma xy=k
def f(n,x):
    return n/x

#Se define la función de la forma xy^(1.4)=k
def g(n,x):
    return (n/x)**(5/7)

#----------------------------------------------------------------------------------
plt.figure(figsize=(8,16))
plt.subplot(211)
#Se grafican las funciones xy=a, xy=b, xy^(1.4)=c y xy^(1.4)=d
plt.plot(x1, f(a,x1), label = 'xy=a')
plt.plot(x2, f(b,x2), label = 'xy=b')
plt.plot(x3, g(c,x3), label = '$xy^{1.4}=c$')
plt.plot(x4, g(d,x4), label = '$xy^{1.4}=d$')

#Añadiendo detalles al gráfico:
plt.ylabel('Presión')
plt.xlabel('Volumen')
plt.title('Gráfica de un ciclo en una Máquina de Carnot')
plt.legend()

#Mostrando el gráfico
plt.show()
```

<font size="4"><br><br>
Aunque el problema se podría abordar al dividir la región $R$ en tres sub-regiones e integral cada región por aparte, la geometría de la región permite que al realizar un cambio de variable para que resulte posible integrar la región en una sola integral más simple.<br>
La transformación aplicada para facilitar el proceso de la integral podría ser:<br><br>
$T: \mathbb{R}^{2}\rightarrow\mathbb{R}^{2} $ <br>
$T\begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} xy \\ xy^{1.4} \end{pmatrix}$

<br>Es decir, se aplica la sustitución $u=xy$ y $v=xy^{1.4}$. Entonces, se puede establecer que la región $D$ en el plano $uv$ la cuál es equivalente a la región $R$ en el plano $xy$ se podría definir matemáticamente como:<br>
$D = \Big\{\begin{pmatrix} u \\ v \end{pmatrix} \in \Pi_{uv} \phantom{a}\big\|\phantom{a} (a \leq u \leq b) \wedge (c \leq v \leq d) \phantom{a}\Big\}$

<br><br><br>A continuación, se presentará la gráfica de la región $D$ en el plano $uv$:

```python
#---------------------------------------------------------------------------------
plt.figure(figsize=(8,16))
plt.subplot(212)
#Se grafican las funciones después de la Transformación:
plt.plot([a,a], [c,d])
plt.plot([b,b], [c,d])
plt.plot([a,b], [c,c])
plt.plot([a,b], [d,d])

#Añadiendo detalles al gráfico:
plt.xlim(a-0.2, b+0.2)
plt.ylim(c-0.2, d+0.2)
plt.ylabel('$xy^{1.4}$')
plt.xlabel('$xy$')
plt.title('Gráfica tras realizar la sustitución $u=xy$ y $v=xy^{1.4}$')

#Mostrando el gráfico
plt.show()
```

<font size="4"><br><br>
Para poder realizar el cambio de variable es necesario calcular el jacobiano de la transformación. Para calcular este, primero se calculará el jacobiano de la transformación inversa $T^{-1}$, el cual estaría definido por:<br><br>
$\dfrac{\partial (u,v)}{\partial (x,y)} = \begin{vmatrix} \frac{\partial u}{\partial x} & \frac{\partial u}{\partial y} 
\\ \frac{\partial v}{\partial x} & \frac{\partial v}{\partial y} \end{vmatrix}
= \begin{vmatrix} y & x \\ y^{1.4} & 1.4xy^{0.4} \end{vmatrix} = 1.4xy^{1.4} - xy^{1.4} = 0.4xy^{1.4} $
<br><br> Dado que $v=xy^{1.4}$, se podría concluir que el jacobiano de la transformación está dado por: <br><br>
$\dfrac{\partial (x,y)}{\partial (u,v)} = \frac{1}{\dfrac{\partial (u,v)}{\partial (x,y)}} = \dfrac{1}{0.4xy^{1.4}} = \dfrac{1}{0.4v} = \dfrac{5}{2v}$
<br><br><br> Por tanto, <br><br>
$W = \displaystyle\iint\limits_{R} dA = \iint\limits_{D} \frac{5}{2v} dA = {5\over2}\int\limits_{a}^{b}\int\limits_{c}^{d}\dfrac{1}{v}dv du = 
{5\over2}\int\limits_{b}^{a}\big[ln(v)\big]_{c}^{d} du = {5\over2}\int\limits_{b}^{a} ln\Big({d \over c}\Big) du 
= {5\over2}ln\Big({d \over c}\Big)\cdot\big[ u \big]_{a}^{b} $
<br><br> $W = \displaystyle{5\over2}(b-a)\cdot ln\Big({d \over c}\Big) $
